# Chicago Crimes Dataset Analysis

## Exploratory Data Analysis (EDA)

### Objectives

This notebook performs exploratory data analysis on the Chicago Crimes dataset.

Goals:

- Understand crime patterns
- Analyze arrest rates
- Explore domestic crimes
- Identify crime hotspots
- Study temporal trends
- Prepare data for dashboard development

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")

plt.rcParams["figure.figsize"] = (10,5)
```

```python
df = pd.read_csv(
    "../data/Crimes_-_2001_to_Present_20260523.csv",
    low_memory=False
)

df.head()
```

```python
print("Rows:", df.shape[0])
print("Columns:", df.shape[1])
```

```python
df.info()
```

Observation:

The dataset contains 22 columns including crime category,
location information, arrest status, and geographical coordinates.

```python
missing = df.isnull().sum()

missing[missing > 0].sort_values(
    ascending=False
)
```

Observation:

Missing values are mainly present in:

- Latitude
- Longitude
- Location
- Location Description

These records will be handled during preprocessing.

```python
# data cleaning
df["Date"] = pd.to_datetime(
    df["Date"],
    format="%m/%d/%Y %I:%M:%S %p",
    errors="coerce"
)

df = df.drop_duplicates()
```

```python
df["Month"] = df["Date"].dt.month
df["Hour"] = df["Date"].dt.hour
df["Day"] = df["Date"].dt.day_name()

df.head()
```

```python
crime_year = df.groupby("Year").size()

crime_year.plot(
    marker="o"
)

plt.title("Crime Trend by Year")
plt.xlabel("Year")
plt.ylabel("Crime Count")

plt.show()
```

Insight:

This chart shows how crime incidents change over time.
Understanding yearly trends helps identify increases or decreases
in criminal activity.

```python
top_crimes = (
    df["Primary Type"]
    .value_counts()
    .head(10)
)

sns.barplot(
    x=top_crimes.values,
    y=top_crimes.index
)

plt.title("Top 10 Crime Types")

plt.show()
```

Insight:

Theft and Battery are among the most common crime categories,
indicating that property and personal crimes dominate the dataset.

```python
crime_pie = (
    df["Primary Type"]
    .value_counts()
    .head(5)
)

plt.pie(
    crime_pie,
    labels=crime_pie.index,
    autopct="%1.1f%%"
)

plt.title("Top 5 Crime Categories")

plt.show()
```

Insight:

A small number of crime categories account for a large
proportion of all reported crimes.

```python
sns.countplot(
    data=df,
    x="Arrest"
)

plt.title("Arrest Distribution")

plt.show()
```

Insight:

Most crimes did not result in an arrest.
This may indicate ongoing investigations or unresolved cases.

```python
sns.countplot(
    data=df,
    x="Domestic"
)

plt.title("Domestic Crime Distribution")

plt.show()
```

Insight:

Domestic incidents represent a smaller portion of all crimes
but remain a significant public safety concern.

```python
districts = (
    df["District"]
    .value_counts()
    .head(10)
)

sns.barplot(
    x=districts.index,
    y=districts.values
)

plt.title("Top Crime Districts")

plt.show()
```

Insight:

Certain districts consistently report higher crime counts,
suggesting geographic concentrations of criminal activity.

```python
locations = (
    df["Location Description"]
    .value_counts()
    .head(10)
)

sns.barplot(
    x=locations.values,
    y=locations.index
)

plt.title("Top Crime Locations")

plt.show()
```

Insight:

Crimes occur most frequently in a limited number of location types,
such as streets, residences, and public areas.

```python
sample = df[
    df["District"].isin(
        df["District"]
        .value_counts()
        .head(10)
        .index
    )
]

sns.boxplot(
    data=sample,
    x="District",
    y="Year"
)

plt.title("District Crime Distribution")

plt.show()
```

```python
sns.violinplot(
    data=sample,
    x="District",
    y="Year"
)

plt.title("District Crime Density")

plt.show()
```

```python
numeric_cols = [
    "Beat",
    "District",
    "Ward",
    "Community Area",
    "Year",
    "Latitude",
    "Longitude"
]

corr = df[numeric_cols].corr()

sns.heatmap(
    corr,
    annot=True,
    cmap="coolwarm"
)

plt.title("Correlation Heatmap")

plt.show()
```

Insight:

The heatmap highlights relationships among numerical variables
and helps identify correlated features.

```python
sample = df.dropna(
    subset=["Latitude","Longitude"]
).sample(
    min(5000, len(df)),
    random_state=42
)

sns.scatterplot(
    data=sample,
    x="Longitude",
    y="Latitude",
    hue="Arrest",
    alpha=0.6
)

plt.title("Crime Location Scatter Plot")

plt.show()
```

Insight:

The geographical distribution reveals concentrations of crime
across specific regions of the city.

```python
monthly = df.groupby(
    "Month"
).size()

monthly.plot(
    kind="bar"
)

plt.title("Monthly Crime Distribution")

plt.show()
```

```python
hourly = df.groupby(
    "Hour"
).size()

hourly.plot(
    kind="line",
    marker="o"
)

plt.title("Crime Distribution by Hour")

plt.show()
```

Insight:

Crime occurrence varies throughout the day,
with certain hours showing noticeably higher activity.
